```python
from sklearn.decomposition import PCA
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import cv2
import matplotlib.pyplot as plt
```

```python
imgRGB = img_to_array(load_img("C:/Study/Semester2/Machine Learning/ML_images/smaller/10.jpg"))
plt.imshow(imgRGB/255)
```

```python
b,g,r = cv2.split(imgRGB)
list_channels = [b,g,r]
    
    #check size first
inverted_img = []
for c in list_channels:            
    pca = PCA(n_components = .99)
    transformed = pca.fit_transform(c/255)
    print(pca.n_components_)
    inverted = pca.inverse_transform(transformed)
    inverted_img.append(inverted)
        
    tuple(inverted_img)
    RGB_compressed = cv2.merge(inverted_img)
```
